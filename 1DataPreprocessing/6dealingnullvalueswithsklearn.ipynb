```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_excel("newdata.xlsx")
```

```python
df.head()
```

```python
from sklearn.impute import SimpleImputer
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df.columns[df.isnull().sum()>0]
```

```python
imputer_mean=SimpleImputer(missing_values=np.nan,strategy='mean')
```

```python
df[['normalized-losses','bore', 'stroke', 'horsepower',
       'peak-rpm', 'price']]=imputer_mean.fit_transform(df[['normalized-losses','bore', 'stroke', 'horsepower',
       'peak-rpm', 'price']]).astype('object')
```

```python
imputer_mode=SimpleImputer(missing_values=np.nan,strategy='most_frequent')
df[['num-of-doors']]=imputer_mode.fit_transform(df[['num-of-doors']])
```
