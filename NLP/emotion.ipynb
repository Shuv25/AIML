```python
f=open("C:\\Users\hp\\Documents\\GitHub\\WORKSPACE\\3AI\\NLP\\read.txt","r",encoding="utf-8").read()

print("Before connverting into lower case : ",end='')
print(f)
```

```python
lc=''
for i in f:
    lc=lc+i.lower()

print("After connverting into lower case : ",end='')
print(lc)
```

```python
print("Before removing special characters : ",end='')
print(lc)
```

```python
nlc=''
special_characters="~!@#$%^&*()_+,.?\/'"
for i in lc:
    if i in special_characters:
        continue
    else:
        nlc=nlc+i
print("After removing special characters : ",end='')
print(nlc)
```

```python
token=nlc.split(' ')
print()
print("Spliting the string into list in short tokenization : ",end='')
print(token)
```

```python
for i in token:
    if i in stop_words:
        token.remove(i)
print()
print("After removing stop words : ",end='')
print(token)
```

```python
fnew = open("C:\\Users\\hp\\Documents\\GitHub\\WORKSPACE\\3AI\\NLP\\emotion.txt").read()

cleaned_text=''
for i in fnew:
    cleaned_text=cleaned_text+i.replace(',','').replace("'",'')

lines=cleaned_text.strip().split('\n')

emotion_dic={}

for line in lines:
    word, emotion = line.split(":")
    word = word.strip()
    emotion = emotion.strip()
    emotion_dic[word]=emotion
print(emotion_dic)
```

```python
emotions=[]
for i in token:
    if i in emotion_dic:
       emotions.append(emotion_dic[i])
print(emotions)
```

```python
count={}
for i in emotions:
    if i in count:
        count[i]+=1
    else:
        count[i]=1

print(count)
```

```python
lll=[]
for i in count:
    lll.append(i)
print(lll)
```

```python
ll=[]

for i in count:
    ll.append(count[i])
print(ll)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8,5))
plt.pie(ll, labels=lll, autopct='%1.1f%%')
plt.show()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8,5))
bars = plt.bar(lll, ll)

for bar, count in zip(bars, ll):
    plt.text(bar.get_x() + bar.get_width() / 2, bar.get_height(), str(count), ha='center', va='bottom')

plt.xlabel('Labels')
plt.ylabel('Counts')
plt.title('Bar Plot with Counts')

plt.xticks(rotation=45)
plt.show()
```
