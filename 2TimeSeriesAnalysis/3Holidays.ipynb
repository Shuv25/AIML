```python
from pandas.tseries.holiday import AbstractHolidayCalendar, nearest_workday,Holiday
class homiesbirthdays(AbstractHolidayCalendar):
    rules=[
        Holiday('Thakur"s birtday',month=2,day=27),
        Holiday('Shuv"s birtday',month=7,day=25),
        Holiday('Kritartha"s birtday',month=11,day=14),
    ]

hol=CustomBusinessDay(calendar=homiesbirthdays())
hol
```

```python
drange=pd.date_range(start='1/1/2024',end='12/31/2024',freq=hol)
```
